```python
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
```

```python
boston_dataset = datasets.load_boston()
```

```python
print(boston_dataset.DESCR)
```

```python
features = dict()
for i, feature_name in enumerate(boston_dataset.feature_names):
    features[feature_name] = boston_dataset.data[:, i]
features['MEDV'] = boston_dataset.target
boston_df = pd.DataFrame(features)
boston_df.head()
```

```python
g = sns.PairGrid(boston_df, y_vars=["MEDV"], x_vars=["PTRATIO", "NOX"], size=4, aspect=1.8)
g.map(sns.regplot, color=".3", )
g = sns.PairGrid(boston_df, y_vars=["MEDV"], x_vars=["LSTAT", "RM"], size=4, aspect=1.8)
g.map(sns.regplot, color=".3")
```
